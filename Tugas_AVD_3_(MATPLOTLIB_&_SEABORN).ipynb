```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/most_used_beauty_cosmetics_products_extended.csv")
df_selected = df.copy()
df
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Rating'], bins=20, kde=True, color='skyblue')
plt.title("Distribusi Rating Produk Kosmetik")
plt.xlabel("Rating")
plt.ylabel("Jumlah Produk")
plt.show()
```

**Berikut adalah visualisasi distribusi rating produk kosmetik, dan insight yang dapat diambil yaitu:**

Sebagian besar produk memiliki rating tinggi di sekitar 4 hingga 5.

Produk dengan rating rendah lebih jarang ditemukan, namun tetap ada.

Distribusi ini menunjukkan bahwa kebanyakan produk kosmetik memiliki kualitas yang cukup baik di mata konsumen.

```python
plt.figure(figsize=(12, 6))
category_avg_rating = df_selected.groupby("Category")["Rating"].mean().sort_values(ascending=False)
sns.barplot(x=category_avg_rating.index, y=category_avg_rating.values, palette="magma")
plt.xticks(rotation=45, ha="right")
plt.title("Rata-rata Rating Berdasarkan Kategori Produk")
plt.xlabel("Kategori Produk")
plt.ylabel("Rata-rata Rating")
plt.show()
```

Produk dengan kategori Contour, Moisturizer, dan CC Cream memiliki rata-rata rating tertinggi, Ini menunjukkan bahwa konsumen cenderung lebih puas dengan produk dari kategori ini, Kategori seperti Lip Liner dan Eyeliner memiliki rata-rata rating lebih rendah dibanding kategori lainnya, Ini bisa menjadi peluang bagi brand kosmetik untuk meningkatkan kualitas atau strategi pemasaran produk dalam kategori ini.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_selected["Price_USD"], y=df_selected["Rating"], alpha=0.5, color="pink")
plt.title("Hubungan antara Harga dan Rating Produk")
plt.xlabel("Harga (USD)")
plt.ylabel("Rating")
plt.xscale("log")  # Menggunakan skala log untuk harga agar lebih terbaca
plt.show()
```

**Visualisasi hubungan antara harga dan rating produk kosmetik:**

Tidak terlihat korelasi yang jelas antara harga dan rating. Produk dengan harga tinggi maupun rendah memiliki variasi rating yang serupa.
Hal ini menunjukkan bahwa faktor lain, seperti kualitas bahan, merek, atau ulasan pengguna, mungkin lebih berpengaruh terhadap rating dibandingkan harga.
Produk dengan harga lebih tinggi tidak selalu mendapatkan rating lebih baik, yang bisa menjadi indikasi bahwa konsumen menilai produk berdasarkan pengalaman penggunaan, bukan sekadar harga.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_selected["Number_of_Reviews"], y=df_selected["Rating"], alpha=0.5, color="blue")
plt.title("Hubungan antara Jumlah Ulasan dan Rating Produk")
plt.xlabel("Jumlah Ulasan")
plt.ylabel("Rating")
plt.xscale("log")  # Menggunakan skala log agar distribusi lebih jelas
plt.show()
```

**Visualisasi hubungan antara jumlah ulasan dan rating produk kosmetik:**

Produk dengan rating tinggi cenderung memiliki lebih banyak ulasan, menunjukkan bahwa produk yang disukai konsumen lebih sering mendapatkan perhatian dan feedback.
Namun, ada juga beberapa produk dengan jumlah ulasan tinggi tetapi memiliki rating rendah, yang bisa mengindikasikan pengalaman negatif pengguna.
Sebagian besar produk dengan rating rendah memiliki jumlah ulasan yang lebih sedikit, yang mungkin menunjukkan kurangnya popularitas atau ketertarikan dari konsumen.

Produk mahal tidak selalu mendapatkan rating lebih tinggi, menunjukkan bahwa faktor lain seperti kualitas dan pengalaman pengguna lebih berpengaruh.

**Hubungan Jumlah Ulasan dengan Rating:** Produk dengan rating tinggi cenderung memiliki lebih banyak ulasan, menandakan tingkat kepuasan yang baik dan kepercayaan konsumen.
